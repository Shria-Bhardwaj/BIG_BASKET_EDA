**Step 8: Find out the outliers from the dataset according to
the columns and fill them with the mean**

```python
q1=df['rating'].quantile(.25)
q3=df['rating'].quantile(.75)
iqr=q3-q1
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
```

```python
lower_limit,upper_limit
```

```python
#extracting clean data
df[(df['rating']>=lower_limit) & (df['rating']<=upper_limit)]
```

```python
q1=df['market_price'].quantile(.25)
q3=df['market_price'].quantile(.75)
iqr=q3-q1
upper_limit=max(0,q3+1.5*iqr)
lower_limit=max(0,q1-1.5*iqr)
```

```python
lower_limit,upper_limit
```

```python
#extracting clean data
df[(df['market_price']>=lower_limit) & (df['market_price']<=upper_limit)]
```

```python
df[df['sale_price']>20000]
```
